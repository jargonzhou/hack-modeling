## 广播

```python
help(np.tile)
```

```python
a = np.arange(0, 40, 10)
print(a)
print(a.shape)  # (4,)
print(np.tile(a, (3, 1)))  # (4,) =prepend 1 dimension=> (1,4) =same size in dimension=> (3, 4)


b = np.array([[0], [1], [2]])
print(b)
print(b.shape)  # 3x1
print(np.tile(b, (4))) # (4) =prepend 1 dimension=> (1, 4) =same size in dimension=> (3, 4)
```
